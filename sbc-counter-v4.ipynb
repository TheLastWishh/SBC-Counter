```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras import layers, models

# =========================
# 0. THAM SỐ CƠ BẢN
# =========================
csv_path = "sbc_counter_result.csv"

n_past = 24                      # số timestep dùng làm input
target_counters = [1, 2, 3]      # các counter_id cần DỰ ĐOÁN
model_type = "transformer"       # "lstm" hoặc "transformer"

# =========================
# 1. LOAD CSV & PIVOT
# =========================
df = pd.read_csv(csv_path)

# Chuyển trigger_time (epoch) -> datetime & sort theo thời gian
df["trigger_time"] = pd.to_datetime(df["trigger_time"], unit="s")
df = df.sort_values("trigger_time")

# Pivot: index = trigger_time, columns = (node_id, counter_id), value = counter_value
pivot = df.pivot_table(
    index="trigger_time",
    columns=["node_id", "counter_id"],
    values="counter_value"
)

# Sắp xếp lại MultiIndex cho chắc chắn
pivot = pivot.sort_index(axis=1, level=[0, 1])

# Lưu lại thông tin về node & counter
node_ids = sorted(df["node_id"].unique())
all_counter_ids = sorted(df["counter_id"].unique())

n_nodes = len(node_ids)
n_counters_total = len(all_counter_ids)

print("Nodes:", node_ids)
print("All counters:", all_counter_ids)

# Flatten columns MultiIndex -> list tên cột dạng chuỗi, vd: "node1_ctr5"
col_pairs = list(pivot.columns)  # list of (node_id, counter_id)
col_names = [f"node{n}_ctr{c}" for (n, c) in col_pairs]
pivot.columns = col_names

# Map từ (node_id, counter_id) -> index cột trong pivot
pair_to_idx = {pair: i for i, pair in enumerate(col_pairs)}

# =========================
# 2. SCALE DỮ LIỆU
# =========================
values = pivot.values  # shape (T, F)
scaler = MinMaxScaler(feature_range=(0, 1))
values_scaled = scaler.fit_transform(values)

T, F = values_scaled.shape
print("Time steps T:", T, "| Features F:", F)

# =========================
# 3. TẠO X, Y CHO BÀI TOÁN
#    - X: (N, n_past, F)
#    - Y: (N, n_nodes, n_target)
# =========================

# Tạo danh sách cột target theo thứ tự: từng node, rồi từng counter_id trong target_counters
target_pairs = []
for node in node_ids:
    for cid in target_counters:
        target_pairs.append((node, cid))

# Index tương ứng trong vector feature dài F
target_indices = [pair_to_idx[p] for p in target_pairs]
n_target = len(target_counters)
out_dim = len(target_indices)  # = n_nodes * n_target

def create_dataset(values_scaled, n_past, target_indices):
    """
    values_scaled: (T, F)
    return:
        X: (N, n_past, F)
        Y: (N, out_dim)  -- sau sẽ reshape thành (N, n_nodes, n_target) nếu cần
        sample_times: (N,) index thời gian tương ứng với bước cần dự đoán
    """
    X, Y = [], []
    times = pivot.index.to_numpy()

    for t in range(n_past, T):
        x_seq = values_scaled[t-n_past:t, :]   # (n_past, F)
        y_vec = values_scaled[t, target_indices]  # (out_dim, )

        X.append(x_seq)
        Y.append(y_vec)

    X = np.array(X)  # (N, n_past, F)
    Y = np.array(Y)  # (N, out_dim)
    sample_times = times[n_past:]  # (N,)

    return X, Y, sample_times

X, Y, sample_times = create_dataset(values_scaled, n_past, target_indices)
N = X.shape[0]
print("X shape:", X.shape, "| Y shape:", Y.shape)

# =========================
# 4. CHIA TRAIN / TEST
# =========================
train_ratio = 0.8
n_train = int(N * train_ratio)

X_train, X_test = X[:n_train], X[n_train:]
Y_train, Y_test = Y[:n_train], Y[n_train:]
time_train, time_test = sample_times[:n_train], sample_times[n_train:]

print("Train samples:", X_train.shape[0], "| Test samples:", X_test.shape[0])

# =========================
# 5. XÂY DỰNG MODEL
# =========================
def build_lstm_model(n_past, F, out_dim, n_nodes, n_target):
    inputs = layers.Input(shape=(n_past, F))  # (B, n_past, F)

    x = layers.LSTM(128, return_sequences=True)(inputs)  # (B, n_past, 128)
    x = layers.LSTM(64, return_sequences=False)(x)       # (B, 64)

    x = layers.Dense(64, activation='relu')(x)           # (B, 64)
    outputs = layers.Dense(out_dim)(x)                   # (B, out_dim)

    # Reshape thành (B, n_nodes, n_target) để dễ xử lý
    outputs_3d = layers.Reshape((n_nodes, n_target))(outputs)

    model = models.Model(inputs=inputs, outputs=outputs_3d)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

def build_transformer_model(n_past, F, out_dim, n_nodes, n_target,
                            d_model=128, num_heads=4, ff_dim=256, num_layers=2):
    inputs = layers.Input(shape=(n_past, F))  # (B, n_past, F)

    # 1) Project features -> d_model
    x = layers.Dense(d_model)(inputs)  # (B, n_past, d_model)

    # 2) Positional embedding (learnable)
    positions = tf.range(start=0, limit=n_past, delta=1)
    pos_embedding_layer = layers.Embedding(input_dim=n_past, output_dim=d_model)
    pos_encoding = pos_embedding_layer(positions)                # (n_past, d_model)
    pos_encoding = tf.expand_dims(pos_encoding, axis=0)          # (1, n_past, d_model)

    x = x + pos_encoding  # (B, n_past, d_model)

    # 3) Transformer encoder blocks
    for _ in range(num_layers):
        # Multi-head self-attention
        attn_input = layers.LayerNormalization(epsilon=1e-6)(x)
        attn_output = layers.MultiHeadAttention(
            num_heads=num_heads,
            key_dim=d_model // num_heads
        )(attn_input, attn_input)
        x = x + attn_output  # residual

        # Feed-forward
        ffn_input = layers.LayerNormalization(epsilon=1e-6)(x)
        ffn_output = layers.Dense(ff_dim, activation='relu')(ffn_input)
        ffn_output = layers.Dense(d_model)(ffn_output)
        x = x + ffn_output  # residual

    # 4) Global pooling theo thời gian
    x = layers.GlobalAveragePooling1D()(x)  # (B, d_model)

    # 5) Dense + Output
    x = layers.Dense(64, activation='relu')(x)       # (B, 64)
    outputs = layers.Dense(out_dim)(x)              # (B, out_dim)
    outputs_3d = layers.Reshape((n_nodes, n_target))(outputs)

    model = models.Model(inputs=inputs, outputs=outputs_3d)
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

n_target = len(target_counters)

if model_type == "lstm":
    model = build_lstm_model(n_past, F, out_dim, n_nodes, n_target)
else:
    model = build_transformer_model(n_past, F, out_dim, n_nodes, n_target)

model.summary()

# =========================
# 6. TRAINING
# =========================
# Lưu ý: model output là (B, n_nodes, n_target), nhưng Y_train hiện là (B, out_dim).
# Ta reshape Y_train thành (B, n_nodes, n_target) cho trùng.

Y_train_3d = Y_train.reshape(-1, n_nodes, n_target)
Y_test_3d  = Y_test.reshape(-1, n_nodes, n_target)

history = model.fit(
    X_train, Y_train_3d,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    verbose=1
)

# =========================
# 7. PREDICT & INVERSE TRANSFORM
# =========================
Y_pred_3d = model.predict(X_test)  # (N_test, n_nodes, n_target)

# Đưa về 2D để map ngược về không gian feature đầy đủ F
Y_test_scaled_2d = Y_test_3d.reshape(-1, out_dim)
Y_pred_scaled_2d = Y_pred_3d.reshape(-1, out_dim)

N_test = Y_test_scaled_2d.shape[0]

# Tạo dummy (N_test, F), gán các cột target = giá trị scaled, sau đó inverse_transform
dummy_true = np.zeros((N_test, F))
dummy_pred = np.zeros((N_test, F))

dummy_true[:, target_indices] = Y_test_scaled_2d
dummy_pred[:, target_indices] = Y_pred_scaled_2d

dummy_true_inv = scaler.inverse_transform(dummy_true)
dummy_pred_inv = scaler.inverse_transform(dummy_pred)

# Lấy lại đúng các cột target
Y_test_inv_flat = dummy_true_inv[:, target_indices]
Y_pred_inv_flat = dummy_pred_inv[:, target_indices]

# Reshape về (N_test, n_nodes, n_target)
Y_test_inv = Y_test_inv_flat.reshape(-1, n_nodes, n_target)
Y_pred_inv = Y_pred_inv_flat.reshape(-1, n_nodes, n_target)

print("Y_test_inv shape:", Y_test_inv.shape)
print("Y_pred_inv shape:", Y_pred_inv.shape)

# =========================
# 8. HÀM VẼ BIỂU ĐỒ
# =========================
def plot_prediction_for_node_counter(
    time_test,
    Y_test_inv,   # (N_test, n_nodes, n_target)
    Y_pred_inv,   # (N_test, n_nodes, n_target)
    node_ids,
    target_counters,
    node_id_to_plot,
    counter_id_to_plot,
    n_samples_to_plot=None
):
    """
    Vẽ biểu đồ cho 1 node & 1 counter trong target_counters
    - time_test: (N_test,) index thời gian (datetime)
    """
    # Tìm index của node và counter trong tensor
    node_idx = node_ids.index(node_id_to_plot)
    ctr_idx = target_counters.index(counter_id_to_plot)

    true_series = Y_test_inv[:, node_idx, ctr_idx]
    pred_series = Y_pred_inv[:, node_idx, ctr_idx]

    if n_samples_to_plot is not None:
        true_series = true_series[:n_samples_to_plot]
        pred_series = pred_series[:n_samples_to_plot]
        times = time_test[:n_samples_to_plot]
    else:
        times = time_test

    plt.figure(figsize=(12, 5))
    plt.plot(times, true_series, label="Thực tế", linewidth=2)
    plt.plot(times, pred_series, label="Dự đoán", linestyle="--")
    plt.xlabel("Thời gian")
    plt.ylabel("Giá trị counter")
    plt.title(f"Node {node_id_to_plot} - Counter {counter_id_to_plot}")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# =========================
# 9. GỌI HÀM VẼ
# =========================
# Ví dụ: vẽ cho node đầu tiên và counter_id = 1
node_id_to_plot = node_ids[0]
counter_id_to_plot = 1

plot_prediction_for_node_counter(
    time_test,
    Y_test_inv,
    Y_pred_inv,
    node_ids,
    target_counters,
    node_id_to_plot=node_id_to_plot,
    counter_id_to_plot=counter_id_to_plot,
    n_samples_to_plot=200  # vẽ 200 điểm cuối, tùy chỉnh
)
```
